#### Optional: Correlation heatmap

- Correlation heatmaps visualize the pairwise correlations between numerical attributes.
- Quickly spot relationships between multiple variables.

**Correlation** refers to a statistical measure that quantifies the extent to which two variables are related or move together. It helps determine if there is a relationship between two variables and the strength and direction of that relationship.

Correlation values fall within the range of -1 to 1:

- A positive correlation (close to 1) indicates that as one variable increases, the other tends to increase as well, signifying a strong positive relationship.
- A negative correlation (close to -1) implies that as one variable increases, the other tends to decrease, suggesting a strong negative relationship.
- A correlation value near 0 indicates a weak or no linear relationship between the variables.

```python
corr_matrix = iris_data.corr()
corr_matrix
```

```python
# Optional: Create a correlation heatmap
correlation_matrix = iris_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
