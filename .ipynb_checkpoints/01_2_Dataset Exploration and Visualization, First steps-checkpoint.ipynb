# Dataset Exploration and Visualization, First steps

## Import the necessary librairies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## A. Loading a Dataset using Pandas

We start by loading the Iris dataset using Pandas' `pd.read_csv()` function.
The dataset is typically available as a CSV file.

```python
# Replace 'dataset.csv' with the actual file path or URL to the Iris dataset.
data = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

## B. Basic Data Exploration with Pandas and NumPy

After loading the dataset, we can perform basic data exploration.
Here are a few common functions:

Display the first few rows of the dataset:

```python
# Display the first few rows of the dataset.
print("First few rows of the dataset:")
print(data.head())
```

Generate summary statistics for numerical columns: 

```python
# Generate summary statistics for numerical columns.
print("\nSummary statistics for numerical columns:")
print(data.describe())
```

```python
# Count unique values in a categorical column (e.g., 'species').
print("\nCounts of each species:")
print(data['species'].value_counts())
```

## C. Visualizing Data Distributions using Matplotlib and Seaborn

We can visualize data distributions to gain insights into the dataset.
In this example, we'll create a boxplot and a histogram for a column.

- Create a boxplot to visualize the distribution of sepal lengths:

```python
# Create a boxplot to visualize the distribution of sepal lengths.
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Distribution of Sepal Lengths by Species')
plt.show()
```

- Create a histogram to visualize the distribution of petal lengths.

```python
# Create a histogram to visualize the distribution of petal lengths.
plt.hist(data['petal_length'], bins=20)
plt.title('Distribution of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```
